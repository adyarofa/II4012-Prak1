# Chatbot Regex Sederhana (Terinspirasi dari ELIZA)

Notebook ini berisi implementasi chatbot sederhana yang menggunakan *regular expressions* (regex) untuk mencocokkan input pengguna dan memberikan respons. Konsep ini terinspirasi dari chatbot klasik bernama ELIZA yang dibuat pada tahun 1960-an.

**Cara Kerja:**
1.  **Pola dan Respons:** Chatbot memiliki daftar pasangan pola regex dan respons yang telah ditentukan.
2.  **Pencocokan:** Ketika pengguna memasukkan kalimat, chatbot akan mencari pola regex yang cocok dengan kalimat tersebut.
3.  **Memberikan Respons:** Jika ditemukan kecocokan, chatbot akan memilih salah satu respons dari daftar yang sesuai dengan pola tersebut. Grup dalam regex (ditandai dengan `()`) dapat digunakan untuk mengambil bagian dari input pengguna dan memasukkannya ke dalam respons.
4.  **Respons Default:** Jika tidak ada pola yang cocok, chatbot akan memberikan respons default.

```python
import re
import random

# Daftar aturan: pola regex dan responsnya
rules = [
    (r'saya butuh (.*)', [
        "Mengapa Anda butuh {0}?",
        "Apakah itu akan benar-benar membantu Anda mendapatkan {0}?",
        "Apakah Anda yakin Anda butuh {0}?"
    ]),
    (r'kenapa kamu tidak bisa (.*)', [
        "Apakah Anda pikir saya tidak bisa {0}?",
        "Mungkin suatu saat saya bisa {0}.",
        "Apakah Anda benar-benar ingin saya bisa {0}?"
    ]),
    (r'saya (.*)', [
        "Mengapa Anda memberitahu saya bahwa Anda {0}?",
        "Sudah berapa lama Anda {0}?",
        "Ceritakan lebih banyak tentang itu."
    ]),
    (r'apakah kamu (.*)', [
        "Menurut Anda, apakah saya {0}?",
        "Mungkin saja, apa pendapat Anda?",
        "Mengapa Anda bertanya apakah saya {0}?"
    ]),
    (r'apa itu (.*)', [
        "Mengapa Anda bertanya tentang {0}?",
        "Apa pendapat Anda tentang {0}?",
        "Sudahkah Anda mencarinya di Google?"
    ]),
    (r'bagaimana (.*)', [
        "Bagaimana menurut Anda?",
        "Mungkin Anda bisa menjawab pertanyaan itu sendiri.",
        "Mengapa Anda bertanya tentang itu?"
    ]),
    (r'(.*) maaf (.*)', [
        "Tidak perlu meminta maaf.",
        "Tidak apa-apa.",
        "Saya mengerti."
    ]),
    (r'hai|halo', [
        "Halo, ada yang bisa saya bantu?",
        "Hai, bagaimana kabarmu?",
        "Senang bertemu denganmu!"
    ]),
    (r'keluar|bye|quit', [
        "Senang berbicara dengan Anda. Sampai jumpa!",
        "Selamat tinggal!",
        "Sampai jumpa lagi!"
    ]),
    (r'(.*)', [
        "Coba katakan itu dengan cara lain.",
        "Saya tidak begitu mengerti.",
        "Ceritakan lebih banyak.",
        "Menarik... lanjutkan.",
        "Lalu apa yang terjadi?"
    ])
]
```

```python
def chatbot_response(user_input):
    """
    Mencari kecocokan input pengguna dengan aturan regex dan mengembalikan respons.
    """
    for pattern, responses in rules:
        match = re.match(pattern, user_input, re.IGNORECASE)
        if match:
            # Memilih respons acak dari daftar
            response = random.choice(responses)
            # Jika ada grup dalam regex, ganti {0}, {1}, dst. dengan grup yang cocok
            if '{0}' in response and match.groups():
                return response.format(match.group(1))
            else:
                return response
    # Fallback, meskipun pola (.*) harusnya menangkap semuanya
    return "Ceritakan lebih banyak."
```

```python
print("Halo, saya adalah chatbot sederhana. Siapa namamu?")
user_name = input("> ")
print(f"Senang bertemu denganmu, {user_name}! Silakan mulai percakapan. Ketik 'keluar' untuk berhenti.")

while True:
    try:
        user_input = input(f"{user_name}> ")
        if not user_input:
            continue
        response = chatbot_response(user_input)
        print(f"Bot> {response}")
        
        # Berhenti jika pengguna mengetik perintah keluar
        if re.match(r'keluar|bye|quit', user_input, re.IGNORECASE):
            break
    except (KeyboardInterrupt, EOFError):
        print("\nBot> Sampai jumpa!")
        break
```
